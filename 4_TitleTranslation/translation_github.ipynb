```
#百度通用翻译API,不包含词典、tts语音合成等资源，如有相关需求请联系translate_api@baidu.com
# coding=utf-8

import http.client
import hashlib
import urllib
import random
import json

# redacted for safety

appid = [redacted]  # 填写你的appid
secretKey = [redacted]  # 填写你的密钥

httpClient = None
myurl = '/api/trans/vip/translate'

fromLang = 'zh'   #原文语种
toLang = 'en'   #译文语种
salt = random.randint(32768, 65536)

def baidu(totrans):
  out = []

  for q in totrans:
    sign = appid + q + str(salt) + secretKey
    sign = hashlib.md5(sign.encode()).hexdigest()
    myurl_full = myurl + '?appid=' + appid + '&q=' + urllib.parse.quote(q) + '&from=' + fromLang + '&to=' + toLang + '&salt=' + str(
    salt) + '&sign=' + sign

    try:
        httpClient = http.client.HTTPConnection('api.fanyi.baidu.com')
        httpClient.request('GET', myurl_full)

        # response是HTTPResponse对象
        response = httpClient.getresponse()
        result_all = response.read().decode("utf-8")
        result = json.loads(result_all)

        print (result)

        # add to array the json file - get the dictionary of trans result, first array item, and the value in dst
        out.append(result['trans_result'][0]['dst'])
        print (out)

    except Exception as e:
        print (e)
    finally:
        if httpClient:
            httpClient.close()
  return out
```
